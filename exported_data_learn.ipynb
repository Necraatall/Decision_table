Analýza vzorku dat pro vytvoření `decision table`.

## Statistická analýza

- Přečtení dat ze souboru `export.csv`
- Analýza vnořených dat a četnosti

```python
import pandas as pd

df = pd.read_csv("export.csv")
df.describe()
```

## Vzorek 20 náhodných záznamů

```python
df.sample(20)
```

## Hodnoty Location

Vybrání sloupečku `Location` a nalezení 20ti nejčastějších záznamů.

```python
from collections import Counter
c = Counter()

for loc in df["Location "]:
    c[loc] += 1

c.most_common(20)
```
